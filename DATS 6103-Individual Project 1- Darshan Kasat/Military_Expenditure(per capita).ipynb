# Per capita Military Expenditure vs Per person GDP


```python
import pandas as pd
pd.set_option('display.precision', 2)
import matplotlib.pyplot as plt
pd.set_option('display.float_format', lambda x: '%.2f' % x)
%matplotlib inline
```

```python
df = pd.read_excel('SIPRI-Milex-data-1949-2016.xlsx', index_col='Country', skiprows=6, sheetname=7, skip_footer=8,
                  na_vales = ['NaN', '.', '. .', 'xxx'])
df.drop(df.columns[0:19], axis=1, inplace = True)
df.dropna(how='any', inplace=True)
df.columns = [2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016]
```

```python
g20_nations = ['Argentina', 'Australia', 'Brazil', 'Canada', 'China, P.R.', 'France', 'Germany',
              'India', 'Indonesia', 'Italy', 'Japan', 'Mexico', 'USSR/Russia', 'Saudi Arabia',
              'South Africa', 'Korea, South', 'Turkey', 'UK', 'USA']
g_20 = df.loc[g20_nations]
g_20.sort_index(inplace=True)
g_20['Average_per_capita(Mili)'] = g_20.mean(axis=1)
g_20
```

```python
df1 = pd.read_excel('GDP-per-person.xlsx', skiprows= 2, index_col='Country')
df1.sort_index(inplace=True)
df1.columns = [2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016]
df1['Average_per_capita(GDP)'] = df1.mean(axis=1)
df1
```

```python
g_20['Average_per_capita(GDP)'] = df1['Average_per_capita(GDP)']
g_20.sort_values(by='Average_per_capita(GDP)', inplace=True, ascending=False)
g_20
```

## Grouped bar chart for Average per capita Milex and Average GDP per person(2006-2016)

```python
plt.style.use('bmh')
g_20.plot(y = ['Average_per_capita(GDP)', 'Average_per_capita(Mili)'], kind='barh',
         figsize = (16,9), fontsize = 13).invert_yaxis()
plt.legend(loc = 'center right')
plt.xlabel('In US Dollars', fontsize = 15)
plt.ylabel('G-20 Nations', fontsize = 15)
plt.title('Average GDP per person v/s Average Milex per person', fontsize = 16)
plt.savefig('comparision.png')
```
